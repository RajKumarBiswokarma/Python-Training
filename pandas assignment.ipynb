```python
# Use the Series.isnull() method to select all rows from f1000 that have a null value for the previous_rank column.
previous_rank_column = f1000[f1000['prev_rank'].isnull()]
previous_rank_column.head()
```

```python
# Select only the company, rank, and previous_rank columns where previous_rank column is null.
# Assign the result to null_previous_rank.
null_previous_rank = f1000.loc[f1000['prev_rank'].isnull(),['company','rank','prev_rank']]
null_previous_rank.head()
```

```python
# Use the Series.notnull() method to select all rows from f1000 that have a non-null value for the previous_rank column. 
# Assign the result to previously_ranked
previous_rank_column = f1000[f1000['prev_rank'].notnull()]
previous_rank_column.head()
```

```python
# From the previously_ranked dataframe, subtract the rank column from the previous_rank column.
# Assign the result to rank_change.
# Assign the values in the rank_change to a new column in the f1000 dataframe, "rank_change".

f1000['prev_rank']=pd.to_numeric(f1000['prev_rank'], errors='coerce')
f1000['rank_change'] = f1000['prev_rank'] - f1000['rank']
f1000.head(6)
```
