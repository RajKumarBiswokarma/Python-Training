```python
cricket_df = pd.read_csv('csv/cricket_scores.csv')
```

```python
cricket_df.head()
cricket_df.info()
```

```python
nz_score = cricket_df.loc[cricket_df['Batting_team'] == 'Newzeland'] 
nz_score['fall_of_wicket'] = nz_score['Wickets'].diff()
nz_score['fall_of_Wicket'] = nz_score['fall_of_wicket'].fillna('0')
nz_score.drop(['fall_of_wicket'],axis =1,inplace = True)
```

```python
nz_scores_fow = nz_score.loc[nz_score['fall_of_Wicket'].astype('int') > 0]
nz_scores_fow
```

```python
fow_dict ={'Run' : [],
          'Overs': [],
          'fow': []}
for item in nz_scores_fow.values:
    fow_dict['Run'].append(item[1])
    fow_dict['Overs'].append(item[2])
    fow_dict['fow'].append(item[4])
```

```python
fow_dict
```

```python
plt.plot(nz_score['Over'],nz_score['Run'] )
plt.scatter(fow_dict['Overs'],fow_dict['Run'])
for i in range(len(fow_dict['fow'])):
    plt.text(fow_dict['Overs'][i], fow_dict['Run'][i] + 5,str(fow_dict['fow'][i]))
plt.title("NewZeland Scores in 2nd T20i")
plt.xlabel("No. of Overs")
plt.ylabel("Run Scores")

```

#### Bike Rental dataset


```python
bike_df = pd.read_csv('csv/bike_rental_details.csv')
bike_df.head()

```

```python
bike_df['dteday'] = pd.to_datetime(bike_df['dteday'])
bike_df.set_index('dteday',inplace = True)
bike_df.head()
```

```python
bike_df_day = bike_df.resample('D').mean()
bike_df_day.head()
```

```python
bike_df_day.reset_index(inplace = True)
bike_df_day.head()
```

```python
plt.plot(bike_df_day['dteday'],bike_df_day['casual'],color = 'green' )# ,label = 'Casual')
plt.plot(bike_df_day['dteday'],bike_df_day['registered'])# label = Registered#)
plt.xticks(rotation = 30)
plt.title("Bikes Rented: Casual vs. Registered")
plt.xlabel("Date")
plt.ylabel('Bikes Rented')
plt.legend(['Casual', 'Registered'])
```

#### Stationary Graph


```python
bike_df_day['Casual_df'] = bike_df_day['casual'].diff()
plt.plot(bike_df_day['dteday'],bike_df_day['Casual_df'],color = 'green' ,label = 'Casual')
plt.xticks(rotation = 30)
plt.title("Casual Data")
plt.xlabel("Date")
plt.ylabel('Bikes Rented')
plt.legend()
plt.xticks(rotation = 30)
plt.show()
```

```python
bike_df_day['Casual_df'] = bike_df_day['casual'].diff()
bike_df_day['casual_df_mv'] = bike_df_day['casual'].rolling(120).mean()
plt.plot(bike_df_day['dteday'],bike_df_day['Casual_df'],color = 'green' ,label = 'Casual')
plt.plot(bike_df_day['dteday'],bike_df_day['casual_df_mv'],color = 'black' ,label = 'Casual')
plt.xticks(rotation = 30)
plt.title("Casual Data")
plt.xlabel("Date")
plt.ylabel('Bikes Rented')
plt.legend()
plt.xticks(rotation = 30)
plt.show()
```

```python
def freq_count(df,col_names,rename_cols):
    
    new_df = df[col_names].value_counts().to_frame().reset_index()
    new_df.columns = rename_cols
    
    return new_df
    
    
```

```python
ceo_founder_df = freq_count(company_df, 'ceo_founder', ['ceo_founder', 'frequency'])
```

```python
ceo_founder_df.head()
```

```python
def bar_plot(df,figsize,x,y,title):
    
    columns = list(df.columns)
    plt.figure(figsize = (figsize))
    plt.bar(df[x], df[y])  # df[columns[0]]
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xticks(rotation = 10)
    plt.show()
    
```

```python
bar_plot(ceo_founder_df,(10,6), 'ceo_founder','frequency','Ceo_founder in Fortune companies')
```

```python
def sns_bar_plot(df,figsize ,x , y, title):
    
    columns = list(df.columns)
    plt.figure(figsize = (figsize))
    plt.title(title)
    ax = sns.barplot(x = df[columns[0]], y =df[columns[1]], label = df[columns[0]])
    plt.legend()
    
    for value in ax.patches:
        ax.text(value.get_x(), value.get_height() + 15, value.get_height())
    plt.show()
    
```

```python
sns_bar_plot(df = owner, figsize = (10,6), x = 'owner', y = 'frequency', title = 'Owner details')
```

```python
fuel_df = freq_count(car_df, 'fuel', ['fuel', 'frequency'])
fuel_df.head()
```

```python
sns_bar_plot(df = fuel_df, figsize =(10,6), x = 'fuel', y = 'frequency', title = 'Fuel Distribution')
```
