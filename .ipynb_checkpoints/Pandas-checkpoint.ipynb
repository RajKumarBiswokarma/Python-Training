```python
# Function for Freq
def freq_table(dataframe, exist_column,new_columns):
    freq_count_df = dataframe[exist_column].value_counts().to_frame().reset_index()
    freq_count_df.columns = new_columns
    return freq_count_df
    
```

```python
seller_count_df = freq_table(dataframe = df_car, exist_column = 'seller_type', new_columns = ['seller_type','freq'])
seller_count_df
```

```python
transmission_count_df = freq_table(dataframe = df_car, exist_column ='transmission', new_columns =['transmission','freq'])
transmission_count_df
```

#####  Handling Missing Values
We can use fillna() method in pandas to fill missing values using different ways.
We can use interpolation method to make a guess on missing values.
We can use dropna() method to drop rows with missing values.
We can also fill missing values with the mean value, median value or the mode value depending on the values of columns.
Filling missing values with mean is appropriate when the column has continuous values.
If the data is categorical then filling missing values with median and mode is a good idea.


```python
df_weather1
```

```python
df_weather1['temperature_fillna_0'] = df_weather1['temperature'].fillna(0)
df_weather1
```

```python
df_weather1['temperature_fillna_mean'] = df_weather1['temperature'].fillna(df_weather1['temperature'].mean())
df_weather1['temperature_fillna_mean'] = df_weather1['temperature_fillna_mean'].round(2)
```
