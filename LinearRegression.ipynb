```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv/adversisement.csv')
df.shape
```

```python
df.head()
```

```python
import seaborn as sns
sns.kdeplot(df['TV'])
```

```python
plt.figure(figsize = (8,6))
plt.scatter(x = df['TV'], y = df['Sales'],color = 'red')
plt.xlabel("TV Advertisement")
plt.ylabel('Sales')
plt.title('Sales for TV advertisement')
plt.show()
```

```python
def mse(y_test,y_pred):
    return np.mean((y_pred - y_test)**2)
```

```python
def mae(y_test,y_pred):
    return np.mean(abs(y_pred - y_test))
```

```python
def rmse():
    return np.mean(np.sqrt((y_pred - y_test)**2))
```

```python
data = df[['TV', 'Radio']]
labels = df['Sales']
```

```python
data.shape
```

```python
labels.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data,labels,test_size = 0.25, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_squared_error(y_pred, y_test, squared = False)
```

```python
mean_absolute_error(y_pred, y_test)
```
